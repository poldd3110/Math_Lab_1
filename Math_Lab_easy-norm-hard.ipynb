$$0 < x_0 < 1 \implies 0 < x_n < 1$$

$0 < x_0 < 1$ <br>
$0 < rx_0(1-x_0) < r(1-x_0)$ <br>
$\begin{rcases}
x_0 < 1 \implies rx_0 < r \\
rx_0 < r \implies r-rx_0 > 0 \\
rx_0 \not= 0 \\
0 < r \leqslant 1 \\
\end{rcases}
\implies 0 < r-rx_0 < 1$ <br>
$0 < rx_0 - rx_0^2 < x_0 \Leftrightarrow 0 < x_1 < x_0 \implies 0 < x_1 < 1$ <br>
По индукции$\space 0 < x_2 < 1, ..., 0 < x_n < 1$

```python
def plot_logistic_map(r_values):
    plt.figure(figsize=(10, 6))

    x = np.linspace(0, 1, 1000)
    for r in r_values:
        y = r * x * (1 - x)
        plt.plot(x, y, label=f"r = {r}", linewidth=2)

    # y=x для сравнения
    plt.plot(x, x, "k--", alpha=0.7, label="y = x", linewidth=2)

    # Настройки графика
    plt.xlabel("$x_{n-1}$", fontsize=14)
    plt.ylabel("$x_n$", fontsize=14)
    plt.title(
        "Логистическое отображение: $x_n = r  x_{n-1} (1 - x_{n-1})$", fontsize=16
    )
    plt.legend(fontsize=12)
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()


plot_logistic_map([1, 1.1, 2, 3, 3.5, 4])
```

Вывод: При бОльших $r$ гораздо сильнее увеличивается влияние $x_i$ которые близки к $0,5$

$\textbf{Normal:} \\$
Для отображения $g(x_n)$, заданного вариантом:
1. Аналитически найдите неподвижную точку.
2. Найдите или оцените диапазон параметра $r$, при котором последовательность монотонно сходится к нулю.
3. Постройте графики зависимости $x_n$ от $n$ для нескольких различных значений параметра $r$.

Неподвижная точка $( x^* )$ удовлетворяет уравнению:<br>
$x^* = f(x^*) = r x^* (1 - (x^*)^2)$<br>
($r = 0$) или ($r = \frac{3\sqrt{3}}{2}$ и $x_i \approx 0.57735$) $\implies x^* = 0$<br>
$0 < r < \frac{3\sqrt{3}}{2}$:<br>
$x^* - r x^* (1 - (x^*)^2) = 0$<br>
$x^* (1 - r (1 - (x^*)^2)) = 0$<br>
$x^* (1 - r + r (x^*)^2) = 0$<br>
$x^* = \sqrt{(1-\frac{1}{r})} \implies 1-\frac{1}{r} \ge 0 \implies r \ge 1$<br>
Получили, что неподвижными точками могут быть только точки $0$ и $\sqrt{(1-\frac{1}{r})}$

Последовательность монотонно сходится к $0$ при $0 < r \le 1$. Докажем:<br>
Если $r = 0$, то $0 = x_1 = x_2 = ...$, поэтому предел равен $0$, но функция не монотонно убывает. Иначе если $0 < r \le 1$:<br>
Рассмотрим разность:<br>
$x_{n+1} - x_n = r x_n (1 - x_n) - x_n = x_n (r (1 - x_n) - 1)$<br>
Так как $x_n \in (0, 1)$, $r \leq 1$, то $r (1 - x_n) - 1 < 0 \implies x_{n+1} - x_n < 0 \implies x_{n+1} < x_n$<br>

Значит, последовательность $\{x_n\}$ монотонно убывает.<br>
Также легко заметить что последовательность ограничена снизу нулём, это значит что по теореме Вейерштрасса существует предел. Обозначим его за $l$:<br>
$\lim_{n \to +\infty} x_n = \lim_{n \to +\infty} rx_{n-1}(1-x^2_{n-1}) \implies l = rl(1-l^2) \implies l(rl^2-r+1) = 0 \implies l = 0$ (второй корень не существует при $0 < r \leq 1$)

Ответ:
При $r \in (0, 1]$ последовательность монотонно убывает и сходится к $0$.<br>

```python
r_values = [0.1, 0.3, 0.5, 0.7, 0.8, 0.9, 1.0, 1.2, 1.4, 1.6, 1.8, 2.0]
x0 = 0.3
n_iter = 100
x = [x0]
plt.figure(figsize=(12, 5))
for r in r_values:
    x = [x0]
    for i in range(n_iter):
        x.append(r * x[-1] * (1 - x[-1]))
    plt.plot(x, "o-", label=f"r = {r}", linewidth=2, markersize=1)
plt.xlabel("$n$", fontsize=14)
plt.ylabel("$x_n$", fontsize=14)
plt.title("Итерации $x_{n+1} = r x_n (1 - x_n^2)$", fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(0, color="k", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

## Hard Level

$\textbf{Hard:}$
1. Положим $r_{\infty} \approx 3.5699456 \ldots$. Как изменяется длина цикла при $r \in (3; r_{\infty})?$
2. Для $r \in (3; r_{\infty})$ экспериментально установите, какие ограничения$^a$ действуют на $m?$

$^a$Здесь имеется в виду не ограниченность сверху или снизу, а то, какую закономерность можно выделить, исследовав изменение длины цикла $m$.

```python
r_values = np.linspace(3.0, 3.5699, 3000)
periods = []
for r in r_values:
    x = 0.5
    for i in range(1000):
        x = r * x * (1 - x)
    trajectory = []
    for j in range(100):
        x = r * x * (1 - x)
        trajectory.append(x)
    found_period = None
    for p in range(1, 51):
        is_periodic = True
        for i in range(p):
            if abs(trajectory[-1 - i] - trajectory[-1 - i - p]) > 0.00001:
                is_periodic = False
                break
        if is_periodic:
            found_period = p
            break
    periods.append(found_period)
plt.figure(figsize=(10, 6))
plt.plot(r_values, periods, 'b.', markersize=3)
plt.xlabel('r')
plt.ylabel('Длина цикла')
plt.title('Как меняется длина цикла при изменении r')
plt.grid(True, alpha=0.3)
plt.ylim(0, 35)
plt.show()
```

Вывод:
1) При изменении параметра r в диапазоне от 3 до r∞ ≈ 3.5699456 наблюдается каскад бифуркаций удвоения периода
2) Возможны только циклы с периодами, равными степеням двойки, длина цикла m стремится к бесконечности

$\textbf{Hard:}$
1. Напишите функцию, которая для заданного параметра $r$ строит лестницу Ламерея.
2. Сделайте выводы: как выглядят циклы различных порядков на графике?

```python
def lamerey_staircase(r, x0=0.5, n_iters=200, n_transient=100, tol=1e-6, detect_cycle=True):
    f = lambda x: r * x * (1 - x)
    xs = np.linspace(0, 1, 500)
    ys = f(xs)
    seq = np.empty(n_iters + 1)
    seq[0] = x0
    for n in range(1, n_iters + 1):
        seq[n] = f(seq[n - 1])
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.plot(xs, ys, 'b', label=r'$y=f(x)=rx(1-x)$')
    ax.plot(xs, xs, 'k--', label=r'$y=x$')
    iters_to_show = n_iters
    x_current = seq[0]
    for i in range(iters_to_show):
        x_next = seq[i + 1]
        ax.plot([x_current, x_current], [x_current, x_next], color='red', lw=0.8)
        ax.plot([x_current, x_next], [x_next, x_next], color='red', lw=0.8)
        x_current = x_next
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'Лестница Ламерея, r={r}, x0={x0}')
    ax.legend()
    ax.grid(True)
    plt.show()
    if detect_cycle:
        tail = seq[n_transient + 1:]
        N = len(tail)
        max_p = min(50, N // 2)
        found_p = None
        for p in range(1, max_p + 1):
            diffs = np.abs(tail[:-p] - tail[p:])
            frac = np.mean(diffs < tol)
            if frac > 0.9:
                found_p = p
                break
        if found_p is None:
            print("Не найден простой период возможно хаос или большой период")
        else:
            print(f"Обнаружен период p = {found_p}")
lamerey_staircase(0.5)
lamerey_staircase(3.2)
lamerey_staircase(3.5)
```

 - Порядок 1 (Неподвижная точка):<br>
 Все линии сходятся в одну точку пересечения параболы и диагонали<br>
 - Порядок > 1:<br>
 Многоугольник с вершинами на параболе и прямой $y = x$, период равен числу повторяющихся паттернов.<br>
 Усложняющийся многоугольник с увеличением r<br>
 При каждом удвоении периода добавляется новая петля<br>
 - Хаос (Бесконечный период):<br>
 Заполнение области без явного цикла, линии никогда не повторяются точно<br>
 Лестница заполняет область между параболой и диагональю<br>

$\textbf{Hard:}$ <br>
Исследуйте: как изменяется длина цикла заданного вариантом отображения $g(x_n)$ с изменением параметра $r$? Постройте соответствующие графики. Есть ли сходства с логистическим отображением?

```python
r_max = 3 * np.sqrt(3) / 2
r_values = np.linspace(0.0, r_max, 3000)
periods = []
for r in r_values:
    x = 0.5
    for i in range(1000):
        x = r * x * (1 - x**2)
    trajectory = []
    for j in range(500):
        x = r * x * (1 - x**2)
        trajectory.append(x)
    found_period = None
    for p in range(1, 51):
        is_periodic = True
        for i in range(p):
            if i < len(trajectory) - p and len(trajectory) > p:
                if abs(trajectory[-1 - i] - trajectory[-1 - i - p]) > 0.00001:
                    is_periodic = False
                    break
        if is_periodic and len(trajectory) > p:
            found_period = p
            break
    periods.append(found_period if found_period is not None else 0)
plt.figure(figsize=(12, 6))
plt.plot(r_values, periods, 'b.', markersize=2, alpha=0.8)
plt.xlabel('r', fontsize=12)
plt.ylabel('Длина цикла', fontsize=12)
plt.title("Длина цикла: $x_n = rx_{n-1}(1-x_{n-1}^2), r \in [0.5, 3√3/2]$", fontsize=14)
plt.grid(True, alpha=0.3)
plt.ylim(0, 35)
plt.xlim(0, r_max)
plt.show()
r_values_bif = np.linspace(0.5, r_max, 3000)
x_vals = []
r_vals = []
for r in r_values_bif:
    x = 0.5
    for i in range(300):
        x = r * x * (1 - x ** 2)
    for j in range(100):
        x = r * x * (1 - x ** 2)
        x_vals.append(x)
        r_vals.append(r)
plt.figure(figsize=(14, 7))
plt.scatter(r_vals, x_vals, s=0.1, color='black', alpha=0.5)
plt.xlabel('Параметр r', fontsize=14)
plt.ylabel('x', fontsize=14)
plt.title('Бифуркационная диаграмма: $x_n = r·x_{n-1}(1-x_{n-1}^2), r \in [0.5, 3√3/2]$',
          fontsize=16, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xlim(0.5, r_max)
plt.ylim(-1.5, 1.5)
plt.show()
```

Выводы:
1) Различия
 - Переход к хаосу отображения $g(x)$ в отличае от логистического отображения
 - После определённого r длина цикла отображения $g(x)$ не всегда равна $2^n$
2) Сходства
 - Каскад бифуркаций удвоения периода при маленьких $r$
 - $x^* = 0$ всегда неподвижна
 - Не более одного цикла
