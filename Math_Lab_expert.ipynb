Инвертируя условие устойчивости получаем условие неустойчивости:<br>
$\exists \varepsilon > 0$  $\forall \delta > 0$ $\exists i$: $|x_0 - x^*| < \delta$, $|x_i - x^*| \ge \varepsilon$.<br>
Конкретно в контексте логистического отображения и $x^* = 0$ Имеем: $\exists \varepsilon > 0$  $\forall \delta > 0$ $\exists i$: $x_0 < \delta$, $x_i \ge \varepsilon$. Докажем это.<br>
Рассмотрим $x_i \le \frac{1}{2}$, тогда $1 - x \ge \frac{1}{2}$. Значит $x_{i+1} = rx_i(1 - x_i) \ge \frac{r}{2}x_i$.<br>
Заметим, что $r > 2$, откуда следует что $\frac{r}{2} > 1$, то есть возрастание последовательности.<br>
Можно обобщить следующим образом: $x_{i + k} \ge (\frac{r}{2})^k * x_i$ при условии, что все $x_i$, $x_{i + 1}$, ..., $x_{i + k - 1} \le \frac{1}{2}$.<br>
Тогда возьмём любой $x_0 < min(\delta, \frac{1}{2})$ и $\varepsilon = \frac{1}{2}$. Тогда из обобщения следует, что $x_i \ge (\frac{r}{2})^i * x_0$.<br> Так как $\frac{r}{2} > 1$, мы имеем неограниченно возрастающую показательную функцию умноженную на $x_0$, значит существует такое наименьшее $i$, что $x_i \ge \frac{1}{2}$, а это именно то что мы хотели доказать, ведь мы как раз взяли $\varepsilon = \frac{1}{2}$.

$\textbf{4 Expert:}$ <br>
Напишите функцию, которая для заданных $x_0, \, y_0 = x_0 + \varepsilon, \, r$ строит две траектории лестницы Ламерея на одном графике. Постройте этот график для $r = 4$. Как можно интерпретировать результаты?

```python
def lamerey_staircase(
    r, x0=0.2, epsilon=0.0001, n_iters=100, show_all_iterations=True, color1='red', color2='blue'):
    f = lambda x: r * x * (1 - x)
    y0 = x0 + epsilon
    x_seq = [x0]
    y_seq = [x0 + epsilon]
    for i in range(n_iters):
        x_seq.append(f(x_seq[-1]))
        y_seq.append(f(y_seq[-1]))
    fig, ax = plt.subplots(figsize=(8, 8))
    xs = np.linspace(0, 1, 1000)
    ys = f(xs)
    ax.plot(xs, ys, 'k', label=r'$f(x) = rx(1-x)$', linewidth=1.5)
    ax.plot(xs, xs, 'k--', label=r'$y = x$', linewidth=1, alpha=0.7)
    def draw_trajectory(seq, color, label, alpha=0.7):
        for i in range(n_iters):
            ax.plot([seq[i], seq[i]], [seq[i], seq[i+1]], 
                   color=color, linewidth=1, alpha=alpha)
            ax.plot([seq[i], seq[i+1]], [seq[i+1], seq[i+1]], 
                   color=color, linewidth=1, alpha=alpha)
        ax.scatter(seq, seq, color=color, s=20, alpha=0.8, label=label)
    draw_trajectory(x_seq, color1, f'$x_0 = {x0:.4f}$')
    draw_trajectory(y_seq, color2, f'$y_0 = {y0:.4f}$ ($\epsilon = {epsilon}$)')
    return x_seq, y_seq
x_seq, y_seq = lamerey_staircase(4)
```

Выводы из графика:<br>
 - Чувствительность к начальным условиям: Даже при очень малом начальном отличии $(\varepsilon = 0.0001)$ траектории быстро расходятся

$\textbf{5 Expert:}$
1. Постройте бифуркационную диаграмму логистического отображения.
2. Проанализируйте: как интерпретировать полученный график?
3. Где на диаграмме находится $r_\infty$? Как ведет себя система до $r_\infty$? После?

```python
r_min = 0
r_max = 4.0
r_values = np.linspace(r_min, r_max, 4000)
x_vals = []
r_vals = []
for r in r_values:
    x = 0.5
    for _ in range(500):
        x = r * x * (1 - x)
    for _ in range(200):
        x = r * x * (1 - x)
        x_vals.append(x)
        r_vals.append(r)

plt.figure(figsize=(14, 7))
plt.scatter(r_vals, x_vals, s=0.1, alpha=0.5)
plt.xlabel('Параметр r', fontsize=14)
plt.ylabel('$x_n$', fontsize=14)
plt.title('Бифуркационная диаграмма логистического отображения\n$x_{n+1}=r x_n(1-x_n)$', fontsize=16)
plt.grid(True, alpha=0.3)
plt.xlim(r_min, r_max)
plt.ylim(0, 1)
plt.axvline(3.5699456, color="red")
plt.show()
```

Выводы:
 - Можно интерпретировать эту диаграмму так что при возрастании r у нас начинается расщепление веток на $ 2 $ части, а после поведение становится хаотическим.
 - До $ r = 1 $ имеем схождение к нулю, до $r = 3$ имеем фактически график $ x = 1 - \frac {1} {r} $, ну а дальше начинается увеличение длин циклов в $ 2 $ раза.
 - И после $ r_\infty $ которое примерно равно 3.6 начинается хаос.
